```python
import numpy as np 
import astropy.units as u  

k = 9 * 10**9 * u.N * u.m**2 /u.C**2
```

```python
q1 = 2 * 10**(-6) *u.C
q2 = -1 * 10**(-6) *u.C  
q3 = 4 * 10**(-6) *u.C 

r31 = 2 *u.m 
r32 = 1 *u.m 
```

```python
F31 = k *(q3 * q1)/(r31**2)

F32 = k *(q3 * q2)/(r32**2)

print(f"F32 = {F32.decompose().value:.3e} {F32.decompose().unit}")
print(f"F31 = {F31.decompose().value:.3e} {F31.decompose().unit}")
Fnet = F31 + F32
print(f"Fnet = {Fnet.decompose().value:.3e} {Fnet.decompose().unit}")
```
